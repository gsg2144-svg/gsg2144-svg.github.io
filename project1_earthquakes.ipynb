##### First let's look at the mean, median, and mode with Pandas:

```python
import pandas as pd

df = pd.read_csv("all_month.csv")
mag = df["mag"].dropna()

print("Pandas")
print("Mean:", mag.mean())
print("Median:", mag.median())
print("Mode:", mag.mode()[0])
```

##### Now, let's try this the hard way, without pandas

```python
import statistics

#Read the CSV file and build lines
f = open("all_month.csv", encoding="utf-8")
data = f.read()
f.close()
data = data.replace("\r\n", "\n")
lines = []
line = ""
for ch in data:
    if ch == "\n":
        lines.append(line)
        line = ""
    else:
        line += ch
if line:
    lines.append(line)

#Build headers
header_chars = lines[0]
header = []
word = ""
for ch in header_chars:
    if ch == ",":
        header.append(word)
        word = ""
    else:
        word += ch
if word:
    header.append(word)

#Find column index for mag
i = None
pos = 0
for h in header:
    if h == "mag":
        i = pos
        break
    else:
        pos += 1

#Build dictionary counting magnitudes
counts = {}
for line in lines[1:]:
    cols = []
    word = ""
    for ch in line:
        if ch == ",":
            cols.append(word)
            word = ""
        else:
            word += ch
    if word:
        cols.append(word)

    #To nos consider if mag has errors in data
    if i is not None and len(cols) > i and cols[i] != "":
        try:
            v = float(cols[i])
        except:
            continue
        counts[v] = counts.get(v, 0) + 1

#Rebuild list of all magnitudes
values = []
for v, freq in counts.items():
    n = 0
    while n < freq:
        values.append(v)
        n += 1

#Calculate statistics
mean = statistics.mean(values)
median = statistics.median(values)
mode = statistics.mode(values)

print("Hard way")
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)


```

##### This is data visualization for the earthquakes of the last 30 days:
