##### First let's look at the mean, median, and mode with Pandas:

```python
import pandas as pd

df = pd.read_csv("all_month.csv")
mag = df["mag"].dropna()

print("Pandas")
print("Mean:", mag.mean())
print("Median:", mag.median())
print("Mode:", mag.mode()[0])
```

##### Now, let's try this the hard way, without pandas

```python
import statistics
import csv

f = open("all_month.csv", encoding="utf-8")
reader = csv.DictReader(f)

df = []
for row in reader:
    if row["mag"] != "":
        df.append(float(row["mag"]))

mean = statistics.mean(df)
median = statistics.median(df)
mode = statistics.mode(df)

print("Hard way")
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

```
