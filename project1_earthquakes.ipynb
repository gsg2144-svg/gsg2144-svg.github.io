```python
import pandas as pd

df = pd.read_csv("all_month.csv")
mag = df["mag"].dropna()

print("Pandas")
print("Mean:", mag.mean())
print("Median:", mag.median())
print("Mode:", mag.mode()[0])
```

```python
values = []
with open("all_month.csv", encoding="utf-8") as f:
    header = f.readline().strip().split(",")
    i = header.index("mag")
    for line in f:
        try:
            v = float(line.split(",")[i])
            values.append(v)
        except:
            pass

values.sort()
n = len(values)

mean = sum(values) / n
median = values[n//2] if n % 2 else (values[n//2 - 1] + values[n//2]) / 2

counts = {}
for v in values:
    counts[v] = counts.get(v, 0) + 1
mode = max(counts, key=counts.get)

print("Hard way")
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
bins = {}
for v in values:
    b = int(v)
    bins[b] = bins.get(b, 0) + 1

print("100 Earthquake counts by magnitude during the past 30 days:")
for b in sorted(bins):
    print(b, ":", "ğŸš©" * (bins[b] // 100))
```
